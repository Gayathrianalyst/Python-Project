##**3) Exploratory Data Analysis (EDA) and Visualization**

###*Campaign Performance*
###➢ Plot a scatter plot to visualize the relationship between Acquisition_Cost and ROI.


```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Acquisition_Cost', y='ROI', data=df_marketing, hue='Campaign_Type', palette='viridis')
plt.xlabel('Acquisition_Cost')
plt.ylabel('ROI')
plt.title('Relationship b/w Acquisition_Cost vs ROI')
plt.xticks(np.arange(1000, 11000, 1500))
plt.yticks(np.arange(0, 3300, 500))
plt.grid(True)
plt.tight_layout()
plt.show()
```

###➢ Create a bar chart to visualize the average Conversion_Rate for different Channel_Used, categorized by Campaign_Type.


```python
plt.figure(figsize=(10,6))
sns.barplot(x='Channel_Used', y='Conversion_Rate', data=df_marketing, hue='Campaign_Type', palette='viridis')
plt.xlabel('Channel_Used')
plt.ylabel('Conversion_Rate')
plt.title('Average Conversion_Rate by Channel_Used and Campaign_Type')
plt.grid(True)
plt.tight_layout()
plt.show()
```

###➢ Visualize the distribution of Engagement_Score across different Campaign_Type using a box plot.


```python
plt.figsize=(10,6)
sns.boxplot(x='Campaign_Type', y='Engagement_Score', data=df_marketing, hue='Campaign_Type', palette='viridis')
plt.xlabel('Campaign_Type')
plt.ylabel('Engagement_Score')
plt.title('Distribution of Engagement_Score by Campaign_Type')
plt.tight_layout()
plt.show()
```

###➢ Analyze the average ROI by Company using a bar chart to compare the profitability of campaigns conducted by different companies.


```python
plt.figsize=(10,6)
sns.barplot(x='Company', y='ROI', data=df_marketing, hue='Company', palette='Set1')
plt.xlabel('Company')
plt.ylabel('ROI')
plt.xticks(rotation=45)
plt.title('Average ROI by Company')
plt.tight_layout()
plt.show()
```

###➢ Examine the correlation between Engagement_Score and Conversion_Rate using a heatmap.


```python
plt.figsize=(10,6)
sns.heatmap(df_marketing.select_dtypes(include=np.number).corr(), annot=True)
plt.title('Correlation b/w Engagement_Score & Conversion_Rate')
plt.tight_layout()
plt.show()
```

##*Customer Segmentation*
###➢ Create a count plot to visualize the distribution of Target_Audience.



```python
plt.figsize=(10,6)
sns.countplot(x='Target_Audience', data=df_marketing, hue='Target_Audience', palette='viridis')
plt.xlabel('Target_Audience')
plt.ylabel('Count')
plt.title('Distribution of Target_Audience')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

###➢ Identify which Customer_Segment has the highest Conversion_Rate for each Language using a bar chart.


```python
plt.figsize=(10,6)
sns.barplot(x='Language', y='Conversion_Rate', data=df_marketing, hue='Customer_Segment', palette='cividis')
plt.xlabel('Language')
plt.ylabel('Conversion_Rate')
plt.title('Conversion_Rate by Customer_Segment and Language')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

###➢ Visualize the distribution of Acquisition_Cost across each Customer_Segment, categorized by Channel_Used, using a box plot.


```python
plt.figsize=(10,6)
sns.boxplot(x='Customer_Segment', y='Acquisition_Cost', data=df_marketing, hue='Channel_Used', palette='viridis')
plt.xlabel('Customer_Segment')
plt.ylabel('Acquisition_Cost')
plt.yticks(np.arange(2000, 10000, 1500))
plt.title('Distribution of Acquisition_Cost by Customer_Segment and Channel_Used')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

###➢ Analyze average Conversion_Rate by Language using a bar chart to compare the effectiveness of campaigns conducted in different languages.


```python
plt.figsize=(10,6)
sns.barplot(x='Language', y='Conversion_Rate', data=df_marketing, hue='Language', palette='RdBu')
plt.xlabel('Language')
plt.ylabel('Conversion_Rate')
plt.title('Average Conversion_Rate by Language')
plt.tight_layout()
plt.show()
```

##*Channel Effectiven*
###➢ Compare the Engagement_Score for different Channels_Used, segmented by Campaign_Type, using a bar chart.


```python
plt.figsize=(10,6)
sns.barplot(x='Channel_Used', y='Engagement_Score', data=df_marketing, hue='Campaign_Type', palette='viridis')
plt.xlabel('Channel_Used')
plt.ylabel('Engagement_Score')
plt.title('Engagement_Score by Channel_Used and Campaign_Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

###➢ Show the distribution of total ROI across different Channels_Used using a pie chart.


```python
plt.figsize=(10,6)
plt.pie(df_marketing.groupby('Channel_Used')['ROI'].sum(), labels=df_marketing['Channel_Used'].unique(), autopct='%1.1f%%')
plt.title('Distribution of ROI by Channel_Used')
plt.ylabel('')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

###➢ Plot a scatter plot to show the relationship between Clicks and Impressions for each Channel_Used.


```python
plt.figsize=(10,6)
sns.scatterplot(x='Clicks', y='Impressions', data=df_marketing, hue='Channel_Used', palette='viridis')
plt.xlabel('Clicks')
plt.ylabel('Impressions')
plt.title('Relationship b/w Clicks and Impressions by Channel_Used')
plt.tight_layout()
plt.show()
```

##*Time-Based Anal*
###➢ Plot the distribution of Duration using a histogram.


```python
plt.figsize=(10,6)
sns.histplot(df_marketing['Duration'], bins=20, kde=True, color='r')
plt.xlabel('Duration')
plt.ylabel('Frequency')
plt.title('Distribution of Duration')
plt.tight_layout()
plt.show()
```

###➢ Analyze how the overall Conversion_Rate has changed over Date for each Company using a line chart.


```python
plt.figsize=(15,9)
sns.lineplot(df_marketing.groupby('Date')['Conversion_Rate'].agg(lambda x: x.mode()[0]), marker='o', color='g', errorbar=None)
plt.xlabel('Date')
plt.ylabel('Conversion_Rate')
plt.title('Overall Conversion_Rate over Date for each Company')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

###➢ Examine the trend of Engagement_Score over Date with a line chart. Geographic Analysis


```python
plt.figsize=(15,9)
sns.lineplot(df_marketing.groupby('Date')['Engagement_Score'].agg(lambda x: x.mode()[0]), marker='o', color='r', errorbar=None)
plt.xlabel('Date')
plt.ylabel('Engagement_Score')
plt.title('Engagement_Score over Date')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

###➢ Determine which location has the highest Acquisition_Cost using a bar chart.




```python
plt.figsize=(10,6)
sns.barplot(x='Location', y='Acquisition_Cost', data=df_marketing, hue='Location', palette='viridis')
plt.xlabel('Location')
plt.ylabel('Acquisition_Cost')
plt.title('Acquisition_Cost by Location')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

###➢ Visualize the Conversion_Rate by different Location, categorized by Target_Audience, using a bar chart.


```python
plt.figsize=(10,6)
sns.barplot(x='Location', y='Conversion_Rate', data=df_marketing, hue='Target_Audience', palette='viridis')
plt.xlabel('Location')
plt.ylabel('Conversion_Rate')
plt.title('Conversion_Rate by Location and Target_Audience')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

###➢ Illustrate the proportion of ROI by Location using a pie chart.

```python
plt.figsize=(10,6)
plt.pie(df_marketing.groupby('Location')['ROI'].sum(), labels=df_marketing['Location'].unique(), autopct='%1.1f%%')
plt.title('Proportion of ROI by Location')
plt.ylabel('')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
